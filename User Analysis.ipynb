```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import rpy2
import nltk
import urllib.request

import cv2

import h5py

from keras.layers import Input, Dense, Lambda, Flatten, Reshape
from keras.layers import Conv2D, Conv2DTranspose
from keras.models import Model
from keras.callbacks import EarlyStopping, ModelCheckpoint
from keras import backend as K
from keras import metrics

from sklearn import svm
from sklearn.cluster import KMeans

import glob
import tqdm
import itertools
np.random.seed(314159)
```

```python
# Read in all data; books, ratings, and read later
books_all = pd.read_csv("good_reads_data/books.csv", delimiter=",", dtype={'isbn': 'str', 'isbn13': 'str'})
ratings_all = pd.read_csv("good_reads_data/ratings.csv", delimiter=",")
read_later = pd.read_csv("good_reads_data/to_read.csv", delimiter=",")
```

```python
# Read in all of the images that had been saved from the exploration
# We will store both the list of book_ids (i.e. the image name) as well
#    as the images themselves as a condensed [0,1] numpy array
jpgs = glob.glob("good_reads_data/constant_size/*.jpg")

book_ids = []
loaded_ims = []

for im in jpgs:
    book_ids.append(int(im.split("/")[-1].split(".")[0]))
    loaded_ims.append(np.array(cv2.imread(im), dtype=np.float16)/255.)
```

```python
# Here we convert from a Python list -> numpy array for ease of use
book_ids = np.array(book_ids)
loaded_ims = np.array(loaded_ims)
```

```python
# This is the VAE taken directly from the Exploration notebook;
# I played around with the latent embedding size, and the number of features;
#   that could be a fun thing to try and see if you can find embeddings that are 
#   large enough to, for instance, capture some semblance of the text in the covers
epsilon_std = 1.0
im_h = 75
im_w = 50
im_ch = 3
latent = 50
features = 16
beta = 1

def sampling(args):
    z_mean, z_log_var = args
    epsilon = K.random_normal(shape=(K.shape(z_mean)[0], latent),
                              mean=0., stddev=epsilon_std)
    return z_mean + K.exp(z_log_var) * epsilon

def Encoder(im_h=75, im_w=50, im_ch=3, latent=5, features=16):
    im_input = Input(shape=(im_h, im_w, im_ch))
    C1 = Conv2D(features, 3, padding='same', activation='relu')(im_input)
    C2 = Conv2D(features, 3, padding='same', activation='relu')(C1)
    C2 = Conv2D(features, 3, padding='same', activation='relu')(C2)
    C2 = Conv2D(features, 3, padding='same', activation='relu')(C2)
    flat = Flatten()(C2)
    H1 = Dense(64, activation='relu')(flat)

    z_mean = Dense(latent)(H1)
    z_log_sd = Dense(latent)(H1)
  
    z = Lambda(sampling)([z_mean, z_log_sd])

    return Model(im_input, [z, z_mean, z_log_sd])

def Decoder(im_h=75, im_w=50, latent=5, features=16):
    latent_input = Input(shape=(latent,))
    H1 = Dense(features*im_h*im_w)(latent_input)
    square = Reshape((im_h, im_w, features))(H1)
    C1 = Conv2DTranspose(features, 3, padding='same', activation='relu')(square)
    C1 = Conv2DTranspose(features, 3, padding='same', activation='relu')(square)
    C1 = Conv2DTranspose(features, 3, padding='same', activation='relu')(square)
    C2 = Conv2DTranspose(im_ch, 3, padding='same', activation='sigmoid')(C1)

    return Model(latent_input, C2)

# Initialize the Models
EncoderModel = Encoder(im_h, im_w, im_ch, latent, features)
DecoderModel = Decoder(im_h, im_w, latent, features)

Im_In = Input(shape=(im_h, im_w, im_ch))
Encoded, z_mean, z_log_sd = EncoderModel(Im_In)
Im_Out = DecoderModel(Encoded)

# Variational Autoencoder
VAE = Model(Im_In, Im_Out)

xent_loss = im_h * im_w * metrics.binary_crossentropy(
            K.flatten(Im_In),
            K.flatten(Im_Out))
kl_loss = - 0.5 * K.sum(1 + z_log_sd - K.square(z_mean) - K.exp(z_log_sd), axis=-1)
vae_loss = K.mean(xent_loss + beta*kl_loss)
VAE.add_loss(vae_loss)

VAE.compile(optimizer='adam', loss=None)
VAE.load_weights("final_50_df.h5")
VAE.summary()
```

```python
# Generate Predictions for all of the images
embedded_space = EncoderModel.predict(loaded_ims)[1]
pred_ims = DecoderModel.predict(embedded_space)
```

```python
# Here we initialize the SVM, as a one-versus-one initialisation

clf = svm.SVC(decision_function_shape='ovo', probability=True)
user_scores = []
unique_nums = []
users = set(ratings_all['user_id'].values) # Get the unique list of user IDs

for user_id in tqdm.tqdm(users):
    # For each user id, we will grab all their ratings and 
    #   filter them based on the books we have covers for
    user_ratings = ratings_all.loc[ratings_all["user_id"]==user_id]
    user_ratings_filtered = user_ratings.loc[user_ratings["book_id"].isin(book_ids)]

    data_y = []
    data_x = []
    for ii, rating in user_ratings_filtered.iterrows():\
        # Here we build up the required data (X: book covers; Y: user rating)
        data_y.append(rating["rating"])
        data_x.append(embedded_space[np.where(book_ids==rating["book_id"])].reshape((50,)))

    data_y = np.array(data_y)
    data_x = np.array(data_x)
    
    locs_ = [np.where(data_y == 1)[0],
             np.where(data_y == 2)[0],
             np.where(data_y == 3)[0],
             np.where(data_y == 4)[0],
             np.where(data_y == 5)[0]] 
    
    
    sel_nums = np.ceil(np.bincount(data_y, minlength=6)/len(data_y)*10) # Generate selection ids
    sel = []

    for ii, num in enumerate(sel_nums):
        if ii > 0 and len(locs_[ii-1]) > 0:
            sel += list(locs_[ii-1][0:int(num)])
    mask = np.ones(data_x.shape[0],dtype=bool) #np.ones_like(a,dtype=bool)
    mask[sel] = False
    
    if (len(set(data_y[~mask])) > 1 and len(data_y[mask]) > 1):
        
        # Fit the model;
        # Generate scores based on naive predictions (i.e. the most common)
        #    and based on fitting with the images
        clf.fit(data_x[~mask], data_y[~mask])
        pic_preds = clf.predict(data_x[mask])
        pic_score = sum(np.array(data_y[mask]) == np.array(pic_preds))/sum(mask)
        pic_score_w_one = sum(abs(pic_preds - data_y[mask]) <= 1)/sum(mask)
        
        naive_pred = np.argmax(np.bincount(data_y[~mask], minlength=6))
        naive_score = sum(data_y[mask]==naive_pred)/sum(mask)
        
        unique_nums.append(len(set(pic_preds)))
        
        user_scores.append({'user_id': user_id, 
                            'pic_score': pic_score,
                            'naive_score': naive_score,
                            'within_one': pic_score_w_one,
                           'unique_nums': len(set(pic_preds))})
    
```

```python
# Convert the scores to a data frame
user_scores_df = pd.DataFrame(user_scores)
```
